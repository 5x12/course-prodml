```python
#encoding

data_encoded = pd.get_dummies(data, columns = ['balcony', 'parking', 'furnished', 'garage', 'storage'], drop_first=True)
data_encoded
```

```python
for i in range(len(data_encoded)):
    if data_encoded.garden[i]=='Not present':
        data_encoded.garden[i]=0
    else: 
        data_encoded.garden[i]=int(re.findall(r'\d+', data_encoded.garden[i])[0])
```

## Defining X and y

```python
X = data_encoded[['area', 
             'constraction_year', 
             'bedrooms', 
             'garden', 
             'balcony_yes', 
             'parking_yes', 
             'furnished_yes', 
             'garage_yes', 
             'storage_yes']]
```

```python
y = data_encoded.rent
```

## Train Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
# Split variables into train and test
X_train, X_test, y_train, y_test = train_test_split(X, #independent variables
                                                    y, #dependent variable
                                                    test_size = 0.2,
                                                    random_state = 3
                                                   )
```

## Building the model

```python
# Fitting Random Forest Regression to the dataset
# import the regressor
from sklearn.ensemble import RandomForestRegressor
 
# create regressor object
rf = RandomForestRegressor(n_estimators=100,
                           random_state=0)
 
# fit the regressor with x and y data
rf.fit(X_train, y_train)

# predict values with the trained model
rf.predict([[85, 2015, 2, 10, 1, 1, 0, 0, 1]])
```

```python
# Return the coefficient of determination of the prediction.

rf.score(X_test,y_test)
```

## Tuning hyperparameters

```python
grid_space={'max_depth':[3,7,12,None],
              'n_estimators':[100, 200, 300, 400],
              'max_features':[3,5,7,9]
#               'min_samples_leaf':[1,2,3],
#               'min_samples_split':[1,2,3]
           }
```

```python
from sklearn.model_selection import GridSearchCV

grid = GridSearchCV(RandomForestRegressor(random_state=0),
                    param_grid=grid_space,
                    cv=5,
                    scoring='r2')

model_grid = grid.fit(X_train,y_train)
```

```python
print(f'Best hyperparameters are {model_grid.best_params_}')
print(f'Best score is {model_grid.best_score_}')
```
